```python
all_paths = list(glob.glob("../csv_data_folder/*.csv"))
column_names_dict = {}
for req_path in all_paths:
    req_file_name = req_path.split("/")[-1].split(".")[0]
    if req_file_name == "game_lineups":
        continue
    else:
        df = pd.read_csv(req_path)
        all_cols = list(df.columns)
        if req_file_name == "game_lineups_cleaned":
            column_names_dict["game_lineups"] = all_cols
        else:
            column_names_dict[req_file_name] = all_cols
```
