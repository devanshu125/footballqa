```python
import pandas as pd
```

```python
def display_data(df):
    print(df.shape)
    display(df.head(10))
    print()
    display(df.tail(10))
    print()
    print(df.info())
```

```python
df = pd.read_csv("../csv_data_folder/game_lineups.csv")
display_data(df)
```

```python
df['number'].unique()
```

```python
# fix number column
df['number'] = df['number'].apply(lambda x: x.replace("\n", "").strip())
# replace "-" with 0 as its not allowed in soccer and hence that will represent number missing and we can make it the same dtype
df['number'] = df['number'].replace("-", 0)
df['number'] = df['number'].astype("int32")

# fix date column
df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d")
display_data(df)
```

```python
# save file
df.to_csv("../csv_data_folder/game_lineups_cleaned.csv", index=False)
```
